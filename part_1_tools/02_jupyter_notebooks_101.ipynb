## 1. What Is a Jupyter Notebook?

### The concept

A traditional Python script (`.py` file) is a sequence of instructions that runs from top to bottom, once, and exits.

A Jupyter notebook (`.ipynb` file) is a **document that combines:**
- **Code cells** — Python (or other languages) that you can run and re-run
- **Markdown cells** — formatted text, equations, images, tables
- **Output** — printed results, plots, tables, displayed inline below each cell

```
┌────────────────────────────────────────────────────┐
│ [Markdown] # Section Title                         │
│ Explain what this code does, with LaTeX if needed  │
├────────────────────────────────────────────────────┤
│ [Code]                                             │
│ In [1]: x = 2 + 2                                  │
│         print(x)                                   │
│ Out[1]: 4           ← output appears here          │
├────────────────────────────────────────────────────┤
│ [Markdown] ## Analysis                             │
│ The result above shows...                          │
└────────────────────────────────────────────────────┘
```

### Why we use notebooks for this course

- **Incremental learning**: run one cell, see the output, understand it, move to the next
- **Images and plots inline**: essential for computer vision — you see results immediately
- **Mixing documentation and code**: each section explains the concept right next to the code
- **Reproducible**: share `.ipynb` files and others can re-run exactly what you did

## 3. Execution Order — The Most Important Concept

### How cells execute

Each cell has a **counter** `[N]` showing its execution order:

```
In [1]:  x = 10          ← ran first
In [2]:  y = 20          ← ran second
In [3]:  print(x + y)   ← ran third → output: 30
```

### The common mistake: out-of-order execution

Here's what goes wrong when you run cells out of order:

```
In [1]:  import numpy as np
In [3]:  x = np.array([1, 2, 3])    ← ran this BEFORE cell 2
In [2]:  import opencv as cv2        ← ran this AFTER cell 3
```

Cell 3 used `np` before it was imported in cell 1... wait, cell 1 was run first, so that's fine. But imagine:

```
Cell A:  model = load_model('weights.h5')  
Cell B:  result = model.predict(image)     ← ran this FIRST by mistake
         # NameError: name 'model' is not defined
```

### The rule

> **Always assume cells must run top-to-bottom.** If something breaks, the first fix to try is: **Kernel → Restart & Run All**.

The `[N]` counter tells you the execution history. If you see `[*]`, the cell is currently running.

## 4. The Kernel — What It Is and When to Restart

### What is the kernel?

The **kernel** is a running Python process that executes your code. Think of it as the "brain" of your notebook:

```
┌─────────────────┐        ┌──────────────────────────────┐
│ Notebook UI     │ ──────▶│ Kernel (Python process)       │
│ (your browser   │ code   │                              │
│  or VSCode)     │◀────── │ - Runs your code              │
│                 │ output │ - Stores all variables        │
└─────────────────┘        │ - Holds imported modules      │
                           └──────────────────────────────┘
```

All variables, imports, and function definitions **persist in the kernel** between cells — until you restart it.

### When to restart the kernel

| Situation | Action |
|---|---|
| Something broke and you don't know why | Kernel → Restart & Run All |
| You imported a module you can't unimport | Restart |
| Memory usage climbed too high | Restart |
| Before submitting/sharing a notebook | Restart & Run All (proves it runs clean) |
| Changed a package version | Restart |
| You deleted a cell that defined something | Restart |

### Kernel menu options

| Option | What it does |
|---|---|
| Restart | Clears all variables, keeps cell outputs |
| Restart & Clear Output | Clears variables AND all cell outputs |
| Restart & Run All | Clears variables, then re-runs every cell top to bottom |
| Interrupt | Stops a currently running cell (useful for infinite loops) |

## 5. Essential Keyboard Shortcuts

### Two modes: Edit mode vs Command mode

- **Edit mode** (green border in Jupyter, cursor visible): you're typing inside a cell
- **Command mode** (blue border, no cursor): you're navigating cells

Switch: `Escape` → command mode | `Enter` → edit mode

### Universal shortcuts (work in both modes)

| Action | Shortcut |
|---|---|
| Run cell, move to next | `Shift + Enter` |
| Run cell, stay in cell | `Ctrl + Enter` |
| Run cell, insert below | `Alt + Enter` |

### Command mode shortcuts (press `Escape` first)

| Action | Shortcut |
|---|---|
| Insert cell **above** | `A` |
| Insert cell **below** | `B` |
| **Delete** cell | `D, D` (press D twice) |
| **Copy** cell | `C` |
| **Cut** cell | `X` |
| **Paste** cell below | `V` |
| Convert to **Markdown** | `M` |
| Convert to **Code** | `Y` |
| **Merge** selected cells | `Shift + M` |
| Select multiple cells | `Shift + Up/Down` |
| Move to top | `Ctrl + Home` |
| Move to bottom | `Ctrl + End` |

### Edit mode shortcuts

| Action | Shortcut |
|---|---|
| Autocomplete | `Tab` |
| Docstring/help popup | `Shift + Tab` |
| Comment/uncomment | `Ctrl + /` |
| Undo | `Ctrl + Z` |

## 6. Magic Commands

Magic commands are special Jupyter commands starting with `%` (line magic) or `%%` (cell magic). They're not Python — they're Jupyter extensions.

| Autosave | Yes | Yes |
| Collaboration | No | Yes (like Google Docs) |

### Displaying images: the key difference

Throughout this course, we use a helper function that works in BOTH environments:

````python
# This helper function is used throughout the course.
# It works identically in VSCode and Colab.

import sys
import matplotlib.pyplot as plt
import numpy as np

IN_COLAB = 'google.colab' in sys.modules

def show_image(img, title="", figsize=(8, 6), cmap=None):
    """
    Display an image inline in both Jupyter/VSCode and Colab.
    
    Handles:
    - BGR (OpenCV) → RGB conversion if 3-channel
    - Grayscale images
    - uint8 and float images
    """
    plt.figure(figsize=figsize)
    
    if img.ndim == 3 and img.shape[2] == 3:
        # Assume BGR (OpenCV format), convert to RGB for display
        img_display = img[:, :, ::-1]  # BGR → RGB by reversing last axis
    else:
        img_display = img
    
    if img_display.dtype == np.uint8:
        plt.imshow(img_display, cmap=cmap)
    else:
        # Float image: clip to [0, 1] range for display
        plt.imshow(np.clip(img_display, 0, 1), cmap=cmap)
    
    if title:
        plt.title(title)
    plt.axis('off')
    plt.tight_layout()
    plt.show()

# Demo: create a simple test image and display it
test_image = np.zeros((200, 400, 3), dtype=np.uint8)
# Add some color blocks
test_image[0:200, 0:133] = [255, 0, 0]    # Blue (BGR) / Red (RGB display)
test_image[0:200, 133:266] = [0, 255, 0]  # Green
test_image[0:200, 266:400] = [0, 0, 255]  # Red (BGR) / Blue (RGB display)

show_image(test_image, title="BGR color test — should show Red | Green | Blue")
````

## 8. Best Practices

### Before sharing or submitting a notebook

Always do: **Kernel → Restart & Run All** to verify the notebook runs clean from top to bottom.

### Managing output size

Long outputs (printing thousands of items, giant arrays) slow down notebooks. Use:

```python
print(array[:5])     # show first 5 elements
print(array.shape)   # show shape, not all values
```

### Cell organization tips

- **One concept per cell** — if a cell is doing 5 different things, split it
- **Name your figures** — `plt.title()` makes it clear what you're looking at
- **Import at the top** — keep all imports in the first code cell
- **Comment your exercises** — future you will thank present you

### Colab-specific tips

- Click the **RAM/Disk** indicator (top-right) to monitor resource usage
- Use **Runtime → Disconnect and delete runtime** to free GPU memory when done
- Add this at the top of each Colab notebook to avoid losing work:

```python
from google.colab import drive
drive.mount('/content/drive')
SAVE_DIR = '/content/drive/MyDrive/pose_workshop/'
```
