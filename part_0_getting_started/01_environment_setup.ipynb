## 1. Why Controlled Environments Matter

### The problem: dependency hell

Imagine you have two projects on your computer:

- **Project A** needs `numpy==1.21` (an older version)
- **Project B** needs `numpy>=1.24` (a newer version)

If you install packages into your **system Python**, you can only have one version of NumPy at a time. Installing the version that Project B needs will **break** Project A.

This is called **dependency hell** and it is one of the most common pain points when working with Python libraries.

### The solution: isolated environments

An **environment** is a separate folder that contains its own Python interpreter and its own set of installed packages. Projects never interfere with each other.

```
System Python            venv_project_A        venv_project_B
─────────────            ──────────────        ──────────────
python 3.11              python 3.10           python 3.11
(no numpy)               numpy 1.21            numpy 1.24
                         opencv 4.7            opencv 4.9
                         torch 1.13            torch 2.1
```

## 2. venv vs Conda — When to Use Which

There are two main tools for creating Python environments:

| Feature | `venv` (built-in) | `conda` (Anaconda/Miniconda) |
|---|---|---|
| Comes with Python | Yes — no install needed | Requires Miniconda or Anaconda install |
| Manages Python version | No — uses system Python | Yes — can install Python 3.8, 3.10, 3.11... |
| Manages non-Python packages | No | Yes — can install CUDA, cuDNN, system libs |
| Package manager | `pip` | `conda` (also supports `pip`) |
| Best for | Pure Python projects | Data science, ML with complex system deps |
| Activation | `venv\Scripts\activate` | `conda activate env_name` |

**Rule of thumb for this course:**
- Parts 1–6 (no deep learning): use `venv` — it's simpler and comes built-in
- Part 7+ (PyTorch, TensorFlow, CUDA): use `conda` — it handles GPU drivers cleanly

## 2b. Docker — The Third Environment Tool (You'll Need It in Part 7)

While `venv` and `conda` solve most environment problems, there is a third tool you will encounter in **Part 7 (EfficientPose)**:

**Docker** creates a completely isolated mini-computer inside your computer — not just an isolated Python environment, but an isolated *operating system + Python + libraries + system dependencies*, all in one package.

| Tool | Isolates | Best for |
|------|----------|----------|
| `venv` | Python packages only | Pure Python projects (Parts 1–6) |
| `conda` | Python + system libs + CUDA | Most ML/DL work (Parts 7–8) |
| **Docker** | Everything (OS, Python, CUDA, drivers) | Old/conflicting environments that `conda` can't fix |

**Why can't `conda` handle everything?**

For most of Part 7, `conda` is enough. But **EfficientPose** (NB 20) was built with TensorFlow 1.14, which is so old that it conflicts at the system level — even `conda` environments have trouble isolating it cleanly. Docker sidesteps this entirely by running the old environment in its own isolated container.

## 4. pip, requirements.txt, and pip freeze

### pip — the Python package installer

```bash
pip install opencv-python          # install latest
pip install opencv-python==4.8.0   # install specific version
pip uninstall opencv-python        # remove
pip list                           # see all installed packages
pip show opencv-python             # details about a package
```

### requirements.txt — your project's dependency list

A `requirements.txt` file lists all packages your project needs, one per line:

```
numpy>=1.24.0
opencv-python>=4.8.0
matplotlib>=3.7.0
```

Install all at once:

```bash
pip install -r requirements.txt
```

### pip freeze — lock your exact versions

After installing and testing, "freeze" the exact versions so collaborators get the same environment:

```bash
pip freeze > requirements_locked.txt
```

This produces something like:
```
numpy==1.24.3
opencv-python==4.8.1.78
matplotlib==3.7.2
...
```

**Tip:** Keep a `requirements.txt` with version ranges (>=) for flexibility, and a `requirements_locked.txt` with exact versions (==) for reproducibility.

## 5. What Is a GPU? CPU vs GPU Analogy

### The factory analogy

Think of computation like a factory producing widgets:

| | CPU | GPU |
|---|---|---|
| Workers | Few (4–32 cores) | Thousands (3,000–10,000 CUDA cores) |
| Each worker | Very fast, very smart, handles complex tasks | Slower, simpler, best at repetitive tasks |
| Best at | Sequential logic, branching, general programs | Massive parallel math (matrix operations) |
| Example | Running your OS, a web browser | Training a neural network |

**Why this matters for computer vision:**

An image is a matrix of numbers. Applying a filter, running a neural network layer, or computing a convolution all boil down to **millions of simple multiplications and additions**. A GPU can do these in parallel and is typically **10–100× faster** than a CPU for this kind of work.

## 6. What Are CUDA and cuDNN?

### CUDA — the language of NVIDIA GPUs

**CUDA** (Compute Unified Device Architecture) is NVIDIA's platform that lets programs send computations to the GPU. It's a software layer that sits between your Python code and the GPU hardware.

```
Your Python code
      ↓
PyTorch / TensorFlow / OpenCV
      ↓
CUDA (NVIDIA's compute platform)
      ↓
GPU hardware (thousands of cores working in parallel)
```

### cuDNN — optimized deep learning primitives

**cuDNN** (CUDA Deep Neural Network library) is a library of GPU-optimized primitives for deep learning: convolutions, pooling, normalization, activation functions. Frameworks like PyTorch and TensorFlow call cuDNN under the hood.

### The CUDA version compatibility chain

This is where many beginners get stuck — everything must be compatible:

```
GPU hardware (e.g., RTX 3080)
  └── NVIDIA Driver (e.g., 535.x)
        └── CUDA Toolkit (e.g., 11.8 or 12.1)
              └── cuDNN (e.g., 8.x)
                    └── PyTorch / TensorFlow (must match CUDA version!)
```

**Example of a working stack:**
- NVIDIA Driver: 535+
- CUDA Toolkit: 12.1
- cuDNN: 8.9
- PyTorch: 2.1.0+cu121

Always check the framework's official install page for the exact compatibility matrix.

### Checking if you have a CUDA-capable GPU

````python
import subprocess
import sys

def check_gpu():
    """Check GPU availability using nvidia-smi and/or torch."""
    # Method 1: nvidia-smi (works without Python packages)
    try:
        result = subprocess.run(
            ['nvidia-smi', '--query-gpu=name,memory.total,driver_version',
             '--format=csv,noheader'],
            capture_output=True, text=True, timeout=5
        )
        if result.returncode == 0:
            print("NVIDIA GPU(s) found via nvidia-smi:")
            for i, line in enumerate(result.stdout.strip().split('\n')):
                name, mem, driver = line.split(', ')
                print(f"  GPU {i}: {name} | Memory: {mem} | Driver: {driver}")
        else:
            print("nvidia-smi not available or no NVIDIA GPU detected")
    except (FileNotFoundError, subprocess.TimeoutExpired):
        print("nvidia-smi not found — NVIDIA GPU or driver may not be installed")

    # Method 2: PyTorch (if installed)
    try:
        import torch
        print(f"\nPyTorch version: {torch.__version__}")
        print(f"CUDA available via PyTorch: {torch.cuda.is_available()}")
        if torch.cuda.is_available():
            print(f"CUDA version: {torch.version.cuda}")
            print(f"cuDNN version: {torch.backends.cudnn.version()}")
            for i in range(torch.cuda.device_count()):
                print(f"  Device {i}: {torch.cuda.get_device_name(i)}")
    except ImportError:
        print("\nPyTorch not installed (needed for Part 7 — skip for now)")

check_gpu()
````

## 8. Running in Google Colab

### What Colab gives you (free tier)

| Resource | Free tier |
|---|---|
| CPU | 2-core Intel Xeon |
| RAM | ~12 GB |
| GPU | NVIDIA T4 (15 GB VRAM) — when available |
| Storage | ~107 GB (non-persistent between sessions) |
| Session limit | ~12 hours, disconnects after ~30 min idle |

### Key differences vs local

| Behavior | Local (VSCode) | Google Colab |
|---|---|---|
| Files persist | Yes | No — lost when session ends |
| File upload | Already on disk | Use `files.upload()` or mount Drive |
| Installing packages | Done once in venv | Must install every session (use `!pip install`) |
| Webcam access | Direct cv2.VideoCapture | Requires JavaScript workaround |
| Displaying images | cv2.imshow() works | Must use matplotlib (cv2.imshow blocked) |

### Mounting Google Drive

To persist files (datasets, calibration files, model weights) across sessions:

```python
from google.colab import drive
drive.mount('/content/drive')

# Now access your files:
# /content/drive/MyDrive/your_folder/your_file.jpg
```

### Enabling GPU in Colab

1. **Runtime** → **Change runtime type**
2. Under **Hardware accelerator** → select **T4 GPU**
3. Click **Save** → the runtime restarts
4. Verify with:

```python
import torch
print(torch.cuda.is_available())   # Should print True
print(torch.cuda.get_device_name(0))  # e.g. 'Tesla T4'
```

### Session management tips

- **Save frequently** to Drive — runtime can disconnect without warning
- **Don't leave idle** — Colab recycles inactive sessions after ~30 minutes
- **Re-run all cells** after reconnecting — state is lost on disconnect
- **Large files**: download to Drive first, then access via mount path
