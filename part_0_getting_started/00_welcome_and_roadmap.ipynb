## 1. What Is 6D Pose?

### The intuition

Imagine you are holding a coffee mug. To describe exactly where and how that mug is positioned in the room you need to answer **two questions**:

1. **Where is it?** — 3 numbers: X, Y, Z coordinates (translation)
2. **How is it oriented?** — 3 numbers: rotation around X, Y, Z axes (rotation)

That's **6 numbers total** → **6 degrees of freedom** → **6D pose**.

```
        Z (up)
        |
        |   / Y (forward)
        |  /
        | /
        +--------> X (right)

  Translation:  (tx, ty, tz)   — where the object is
  Rotation:     (rx, ry, rz)   — how the object is turned
```

<img src="https://lh3.googleusercontent.com/d/18-h_x8XUBiL6qFL56wyzyNoglXCR5kmX" width="420" alt="6D pose illustration — translation and rotation of an object in 3D space"/>

### The math (preview — full detail in Part 3)

The 6D pose of an object relative to a camera is represented as:

$$\mathbf{T} = \begin{bmatrix} \mathbf{R} & \mathbf{t} \\ \mathbf{0}^T & 1 \end{bmatrix} \in \mathbb{R}^{4 \times 4}$$

Where:
- $\mathbf{R} \in \mathbb{R}^{3 \times 3}$ is the **rotation matrix** — encodes orientation (3 degrees of freedom)
- $\mathbf{t} \in \mathbb{R}^{3}$ is the **translation vector** — encodes position (3 degrees of freedom)
- The full $4 \times 4$ matrix is called a **homogeneous transformation matrix** (it lets you chain transforms with matrix multiplication)

### What 6D pose looks like in practice

When a camera sees an object, 6D pose estimation tells you:

> "The pallet is **1.2 m in front** of the camera, **0.05 m to the left**, **0.3 m above ground**, and **rotated 15° clockwise** around the vertical axis."

That is all the information a robot needs to reach out and grab it.

## 2. Why Robotics Needs 6D Pose

### The core problem robots face

A robot's **arm, gripper, or entire body** needs to reach a precise position in space. Without vision, robots rely on pre-programmed fixed positions — which breaks the moment anything shifts even slightly.

**6D pose estimation makes robots adaptive:**

| Scenario | What the robot needs to know | Method we'll use |
|---|---|---|
| Mobile robot docking at a charging station | "How far off-center am I? How much do I need to rotate?" | ArUco marker on station wall |
| Forklift picking a pallet | "Where is the pallet? Are the forks aligned with the slots?" | CAD-based 6D pose estimation |
| Conveyor station handoff | "Is the part at the correct angle for my gripper?" | ArUco or deep learning |
| Surgical robot arm | "Exactly where is the tool tip relative to tissue?" | Stereo vision + RGB-D |

### Three real applications in this course

## 6. The Two Real-World Problems We Will Solve

Keep these in mind as you learn — every technique in the course is a building block toward one of them.

### Problem 1 — Autonomous Docking at a Charging / Conveyor Station

**Scenario:** A mobile robot needs to navigate to a fixed docking station, align precisely, and dock — without human guidance.

**Challenge:** The robot's dead-reckoning (wheel odometry) accumulates drift. After 10 minutes of operation, it could be 5–10 cm off and 3–5° misaligned — not acceptable for a physical docking connector.

**Vision solution:**
1. Mount an **ArUco marker** on the station wall
2. Camera on the robot detects the marker
3. Estimate the marker's **6D pose** relative to the camera
4. Compute the **correction offset**: how much to move laterally and rotate
5. Execute the correction, dock, done

**Building blocks we'll need:**
- Part 2: Read and process camera frames
- Part 3: Camera intrinsics (needed for pose from marker corners)
- Part 5: ArUco detection and pose estimation
- Part 8: Full pipeline with correction logic

---

### Problem 2 — Forklift Pallet Detection and Fork Alignment

**Scenario:** A forklift (autonomous or semi-autonomous) needs to slide its forks into the two openings of a wooden pallet for pick-up.

**Challenge:** Pallets are never placed perfectly. They can be rotated, tilted, at varying distances. The forks have only ~1 cm of clearance on each side.

**Vision solution:**
1. A camera is mounted on the forklift mast
2. A **CAD model** of the pallet is used as reference
3. **6D pose estimation** locates the pallet and its orientation
4. The system computes the exact position and angle of the pallet openings
5. The forklift adjusts its approach trajectory accordingly

**Building blocks we'll need:**
- Parts 2–4: Image processing + classical pose estimation
- Part 7: Deep learning for marker-free detection
- Part 8: Full pipeline with CAD model + alignment logic
